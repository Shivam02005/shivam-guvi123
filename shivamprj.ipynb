```python
def get_student_data():
    """Gathers student names and marks from the user."""
    students = {}
    while True:
        name = input("Enter student name (or type 'done' to finish): ")
        if name.lower() == 'done':
            break
        try:
            marks1 = int(input(f"Enter marks for subject 1 for {name}: "))
            marks2 = int(input(f"Enter marks for subject 2 for {name}: "))
            marks3 = int(input(f"Enter marks for subject 3 for {name}: "))
            students[name] = [marks1, marks2, marks3]
        except ValueError:
            print("Invalid input. Please enter numeric marks.")
    return students

def calculate_grades(students):
    """Calculates total, average, and assigns a grade for each student."""
    student_summary = {}
    for name, marks in students.items():
        total = sum(marks)
        average = total / len(marks)

        # Grade assignment logic
        if average >= 90:
            grade = 'A'
        elif average >= 80:
            grade = 'B'
        elif average >= 70:
            grade = 'C'
        elif average >= 60:
            grade = 'D'
        else:
            grade = 'F'

        student_summary[name] = {'total': total, 'average': average, 'grade': grade}
    return student_summary

def print_summary(student_summary):
    """Prints the summary for each student and the class."""
    if not student_summary:
        print("\nNo student data to summarize.")
        return

    print("\n--- Student Summary ---")
    for name, data in student_summary.items():
        print(f"Student: {name}")
        print(f"  Total Marks: {data['total']}")
        print(f"  Average Marks: {data['average']:.2f}")
        print(f"  Grade: {data['grade']}")
        print("-" * 20)

    # Calculate class average and find the topper
    total_class_marks = sum(data['total'] for data in student_summary.values())
    class_average = total_class_marks / len(student_summary)
    
    topper = max(student_summary, key=lambda name: student_summary[name]['total'])

    print("\n--- Class Insights ---")
    print(f"Class Average Score: {class_average:.2f}")
    print(f"Class Topper: {topper} with {student_summary[topper]['total']} marks! üèÜ")

# Main program execution
if __name__ == "__main__":
    student_marks = get_student_data()
    summary_data = calculate_grades(student_marks)
    print_summary(summary_data)
```
